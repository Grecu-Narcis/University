```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
computer_hardware = fetch_ucirepo(id=29) 
  
# data (as pandas dataframes) 
X = computer_hardware.data.features
y = computer_hardware.data.targets
  
# metadata 
print(computer_hardware.metadata) 
  
# variable information 
print(computer_hardware.variables) 
print(computer_hardware.data.targets)
```

Perform a regression using ANNs on the 1-8 features and compare your results with each of the target values. 


```python
y = X[['PRP']]
X = X.drop(['PRP', 'ERP'], axis=1)
# X = X.drop(['VendorName'], axis=1)
X.head()
```

```python
y.head()
```

```python
# encode and normalize the data
from sklearn.preprocessing import LabelEncoder

model_encoder = LabelEncoder()
vendor_encoder = LabelEncoder()
X['ModelName'] = model_encoder.fit_transform(X['ModelName'])
X['VendorName'] = vendor_encoder.fit_transform(X['VendorName'])
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()

for column in ['MYCT', 'MMIN', 'MMAX', 'CACH', 'CHMIN', 'CHMAX']:
    X[column] = min_max_scaler.fit_transform(X[[column]])
    
X.head()
```

```python
# split the data
# X = X.astype(float)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
X_train
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from collections import OrderedDict
torch.manual_seed(42)
# i create a model with 8 input features and 3 hidden layers
model = nn.Sequential(OrderedDict([
    ('hidden1', nn.Linear(8, 100)),
    ('ReLU1', nn.ReLU()),
    ('hidden2', nn.Linear(100, 100)),
    ('ReLU2', nn.ReLU()),
    ('hidden3', nn.Linear(100, 100)),
    ('ReLU3', nn.ReLU()),
    ('output', nn.Linear(100, 1))
]));

print(model)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(params=model.parameters(),lr=0.01)
```

```python
# training
losses = []
X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
y_train_prp_tensor = torch.tensor(y_train.values, dtype=torch.float32)

for epoch in range(10000):
    optimizer.zero_grad()
    outputs = model(X_train_tensor)
    
    loss = criterion(outputs, y_train_prp_tensor)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())

    predicted_classes = (outputs.round() == y_train_prp_tensor)
    accuracy = predicted_classes.sum().item() / len(y_train_prp_tensor)

    if (epoch + 1) % 100 == 0:
        print(f'Epoch: {epoch + 1}, loss: {loss.item()}')
        print(f'Accuracy: {accuracy}')
        # print(outputs)
        # print(y_train_prp_tensor)

    if accuracy == 1:
        print(f'Model reached 100% accuracy at epoch {epoch + 1}')
        break
# print(outputs)
print('done')
```
